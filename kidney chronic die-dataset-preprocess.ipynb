# numerical column------>take mean

```python
import pandas as pd
dataset = pd.read_csv('kidney_disease.csv')
print(dataset.head())
print(dataset.dtypes)
numerical_columns = dataset.select_dtypes(include=['float64', 'int64']).columns
mean_values = dataset[numerical_columns].mean()
#Print the mean values for each numerical column
print("\nMean values for numerical columns:")
print(mean_values)
```

```python
#Fill missing values in numerical columns with the mean
dataset[numerical_columns] = dataset[numerical_columns].fillna(dataset[numerical_columns].mean())

#Check again for missing values
missing_values_numerical_after_fill = dataset[numerical_columns].isnull().sum()

print("\nMissing values after filling with mean:")
print(missing_values_numerical_after_fill)
```

# here we drop the missing values------>'pcv','rc','wc'

```python
dataset.dropna(axis=1)
```

# categorical column--------->take--->Mode

```python
categorical_columns = ['rbc','pc','pcc','ba','htn','dm','cad','appet','pe','ane']  # Add other categorical columns here
#Fill missing values in all categorical columns with the mode
for col in categorical_columns:
    dataset[col] = dataset[col].fillna(dataset[col].mode()[0])
#Check the result
print("\nUpdated DataFrame with missing values filled:")
print(dataset)
```
