#How to split column Quan & Qual------>logic check

```python
quan=[]
qual=[]
for columnname in dataset.columns:
    print(columnname)
    if(dataset[columnname].dtype=='O'):
        print("qual")
        qual.append(columnname)
    else:
        print("quan")
        quan.append(columnname)
            
        
```

```python
quan
```

```python
qual
```

# convert the logic to function

```python
def quanQual(dataset):
    quan=[]
    qual=[]
    for columnname in dataset.columns:
        print(columnname)
        if(dataset[columnname].dtype=='O'):
            print("qual")
            qual.append(columnname)
        else:
            print("quan")
            quan.append(columnname)
    return quan,qual
            
        
    
```

```python
quanQual(dataset)
```

```python
quan,qual=quanQual(dataset)
```

# Median column ------------>values assign

```python
descriptive=pd.DataFrame(index=["Mean","Median","Mode"],columns=quan)
for columnName in quan:
    descriptive[columnName]["Mean"]=dataset[columnName].mean()
    descriptive[columnName]["Median"]=dataset[columnName].median()
```

# Mode column----------->values assign

```python
descriptive=pd.DataFrame(index=["Mean","Median","Mode"],columns=quan)
for columnName in quan:
    descriptive[columnName]["Mean"]=dataset[columnName].mean()
    descriptive[columnName]["Median"]=dataset[columnName].median()
    descriptive[columnName]["Mode"]=dataset[columnName].mode()
```

# Mode column-------->mode()[0]---------------->we get the result----------whole value example---->062.0
